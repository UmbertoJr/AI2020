## Gray encode and Decode functions

```
import numpy as np
from AI2020.DeJong import De_Jong

example = De_Jong(1)
print("DEC    =>    GRAY    => DEC")
numbers = np.around(np.linspace(example.range[1]-8*example.resolution_factor, example.range[1], 9), 2)
for i in numbers:
  gray = example.gray_encode(i)
  dec = example.gray_decode(gray)
  print(f" {i:.2f}  => {gray} => {dec}")
```

```
#example no gray encode
population = np.array([[-5.12, -5, -4], [0, 0, 0], [1, -2, -3]])
print(population)
print(example.evaluate(population, gray_=False))
```

```
#example with gray encode
population_string = []
for i in range(population.shape[0]):
  new_ind = []
  for j in range(population.shape[1]):
    new_ind.append(example.gray_encode(population[i][j]))
  population_string.append(new_ind)
print(np.array(population_string))
print(example.evaluate(population_string, gray_=True))
```
